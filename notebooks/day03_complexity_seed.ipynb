```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 100
field = np.random.randint(0, 2, (size, size))
```

```python
# Glider (a pattern that "moves" in Conway's Game of Life)
glider = np.array([[0, 1, 0],
                   [0, 0, 1],
                   [1, 1, 1]])

# Inject it at position (10, 10)
x, y = 10, 10
field[x:x+3, y:y+3] = glider
```

```python
plt.imshow(field, cmap='binary')
plt.title("Initial Field with Glider")
plt.axis('off')
plt.show()
```

```python
def evolve(field):
    size = field.shape[0]
    next_field = np.zeros_like(field)
    
    for i in range(1, size - 1):
        for j in range(1, size - 1):
            neighbors = np.sum(field[i-1:i+2, j-1:j+2]) - field[i, j]
            if field[i, j] == 1:
                next_field[i, j] = 1 if neighbors in [2, 3] else 0
            else:
                next_field[i, j] = 1 if neighbors == 3 else 0
    return next_field
```

```python
steps = 20
for step in range(steps):
    field = evolve(field)
    
    if step % 5 == 0:  # Plot every 5 steps
        plt.imshow(field, cmap='binary')
        plt.title(f"Step {step}")
        plt.axis('off')
        plt.show()
```
